```python
def master_workflow(seed=None, odim=None, dim_reduct_algo=None, dset=None, 
                        task=None, head_type=None, met=None):
    '''
    A seeded single-iteration workflow of data loading, kernel calculation, 
        dimensionality reduction, model fit and evaluate
    '''

    assert seed is not None, "provide seed!"
    assert odim is not None, "provide output dimension!"
    assert dim_reduct_algo is not None, "provide dimension reduction algorithm!"
    
    # set seed
    seed_everything(seed)

    # load data
    if dset == 'fmnist':
        x_train, y_train, y_train_oh, x_test, y_test = load_fmnist(sample_sz=0.1)
    else:
        x_train, y_train, y_train_oh, x_test, y_test = load_skl_dset(dset, task=task)

    # kernel calculation with gpytorch
    B, K, K_test, n = get_kerns(x_train, x_test, oh_labels=y_train_oh, task=task)

    # dimensionality reduction
    if dim_reduct_algo == 'pca':
        z_train, z_test = pca(x_train, x_test, odim)
    elif dim_reduct_algo == 'spca':
        z_train, z_test = spca(x_train, x_test, B, odim)
    elif dim_reduct_algo ==  'kspca':
        z_train, z_test = kspca(K, K_test, B, odim)
    elif dim_reduct_algo == 'srp':
        z_train, z_test = srp(x_train, x_test, y_train_oh, output_dim=odim)
    elif dim_reduct_algo == 'ksrp':
        z_train, z_test = ksrp(K, K_test, y_train_oh, output_dim=odim)
    else:
        raise NotImplementedError('dimension reduction algorithm not implemented yet!')

    # model fit and evaluate
    result = fit_and_eval(z_train, y_train, z_test, y_test, met=met, 
                          head_type=head_type, verbose=0)
    
    return result
```

```python
def multi_trial(num_trials=10, odim=2, dim_reduct_algo=None, dset=None, 
                task=None, head_type=None, met=None):

    num_trials = 10
    results = np.zeros(num_trials)

    for i in range(num_trials):
        res = master_workflow(seed=i, odim=odim, dim_reduct_algo=dim_reduct_algo, dset=dset,
                            task=task, head_type=head_type, met=met)
        results[i] = res

    mean_result = np.mean(results)
    
    return mean_result
```

```python
max_out_dim = 10
out_dims = np.arange(1, max_out_dim+1)
DSET = 'wine'
TASK = 'class'
HEAD = 'logreg'
METRIC = 'accuracy'

algos = ['pca', 'spca', 'kspca', 'srp', 'ksrp']
res_mat = np.zeros((len(algos), max_out_dim))

for j, algo in enumerate(algos):
    algo_results = np.zeros(max_out_dim)
    for i, dim in enumerate(out_dims):
        res = multi_trial(odim=dim, dim_reduct_algo=algo, dset=DSET, task=TASK,
                   head_type=HEAD, met=METRIC)
        algo_results[i] = res
    res_mat[j] = algo_results
    
res_mat = res_mat.transpose()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='retina'

# change result array into a df for sns plot 
res_df = pd.DataFrame(res_mat, columns=algos, index=out_dims)
res_df['dimensions'] = out_dims

sns.set_theme()
sns.set(font_scale=1.2)

for algo in algos:
    plot = sns.lineplot(x='dimensions', y=algo, data=res_df, lw=2.5)
plt.legend(labels=algos, facecolor='white')
plot.set(ylabel=METRIC, title = f'Dim reduction algos on {DSET} with {HEAD}, by dimensions');
```
